```python
import os
import cv2
import numpy as np
from PIL import Image
from tensorflow.keras.utils import img_to_array 
import pickle
from google.colab.patches import cv2_imshow
```

```python
default_image_size = tuple((256, 256))

def convert_image_to_array(image_dir):
    try:
        image = cv2.imread(image_dir)
        if image is not None :
            image = cv2.resize(image, default_image_size)   
            return img_to_array(image)
        else :
            return np.array([])
    except Exception as e:
        print(f"Error : {e}")
        return None
```

```python
directory = '/content/drive/MyDrive/plant disease classification/testing/pics'

images_list = os.listdir(directory)

images = []

for image in images_list:
    loc = f"{directory}/{image}"
    im = convert_image_to_array(loc)
    images.append(im)
    
#image normalisation
np_image_list = np.array(images, dtype=np.float16) / 225.0    
print(np_image_list.shape)
```

```python
from google.colab.patches import cv2_imshow
for i in np_image_list:

  cv2_imshow(i*255)
```

```python
model = pickle.load(open('/content/drive/MyDrive/plant disease classification/zz/cnn_model.pkl', 'rb'))
```

```python
for i in range(24):
  im = np_image_list[i]
  im = im.reshape(1,256,256,3)
  result = model.predict(im)
  result=np.argmax(result)

  if result==0:
    print("class= bp")
  elif result==1:
    print("class= hispa")
  elif result==2:
    print("class= healthy")
  else:
    print("class= leaf blast")




```
